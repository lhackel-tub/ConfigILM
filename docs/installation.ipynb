# Installation

This sections covers how to get started with `ConfigVLM`.

There are currently two ways of installing `ConfigVLM`. The recommended way is using [pypi](https://pypi.org/) via

```python
# skip-execution
!pip install configvlm
```

Alternatively, the source can be directly downloaded from [github](https://github.com/lhackel-tub/ConfigVLM). If you want to use the extras (e.g. Datasets), installation is `pip install configvlm[extra]`.

:::{note}
Using the extras means installing some additional dependencies.
:::

:::{note}
If you use the wheel, installation would equate to `pip install configvlm-0.1.0-py3-none-any.whl` and `pip install configvlm-0.1.0-py3-none-any.whl[extra]`
:::

`ConfigVLM` allows you to easily combine and use predefined vision and language models and use them for tasks such as [Supervised Pretraining](sup_pretraining.ipynb) or [VQA](vqa.ipynb).
For this, models can be trained in an **end-to-end** fashion or **pre-trained checkpoints** can be used. How the configuration is set up is explained on the next page. Alternatively, use cases for this framework are presented on the following pages.

For all examples in this documentation it is assumed that `ConfigVLM` has been installed together with the `[extra]` dependencies.

If you want to try out the framework's functionality, check out the [google colab](https://colab.research.google.com/drive/1ALBLQdn_NKd0ve7gwJtuNaUU3PgxEbMa?usp=sharing), which shows an example use case.
